###DDL 1 - High Rated Listings

```python
%sql
CREATE OR REPLACE TABLE airbnb_high_rated AS
SELECT 
    Name,
    `Neighbourhood ` AS Neighbourhood,
    Price,
    `Review Scores Rating` AS Review_Scores
FROM airbnb
WHERE `Review Scores Rating` > 85;

SELECT Name, Neighbourhood, Price, Review_Scores
FROM airbnb_high_rated
LIMIT 5;
```

###DDL 2 - Mid Price Range

```python
%sql
CREATE OR REPLACE TABLE airbnb_mid_price AS
SELECT 
  Name,
  `Neighbourhood ` AS neighbourhood,
  Price
FROM airbnb
WHERE Price BETWEEN 100 AND 300;

SELECT Name, neighbourhood, Price
FROM airbnb_mid_price
LIMIT 5;
```

###DDL 3 - Price Categorization

```python
%sql
UPDATE airbnb
SET price_category = CASE
    WHEN Price < 100 THEN 'Budget'
    WHEN Price BETWEEN 100 AND 300 THEN 'Mid-range'
    WHEN Price > 300 THEN 'Luxury'
    ELSE 'Unknown'
END
WHERE price_category IS NULL OR price_category = '';
SELECT Name, Price, price_category
FROM airbnb
WHERE price_category IS NOT NULL
ORDER BY Price
LIMIT 10; 
```

###DDL 4 - Group by Neighbourhood

```python
%sql
CREATE OR REPLACE TABLE airbnb_neighbourhood_summary AS
SELECT 
    `Neighbourhood ` AS Neighbourhood,
    COUNT(*) AS total_listings
FROM airbnb
GROUP BY `Neighbourhood `;


-- Display result
SELECT 
    Neighbourhood,
    total_listings
FROM airbnb_neighbourhood_summary
ORDER BY total_listings DESC
LIMIT 5;
```

###DDL 5 - Sorting by Price

```python
%sql
CREATE OR REPLACE TABLE airbnb_sorted AS
SELECT 
    Name,
    `Neighbourhood ` AS Neighbourhood,
    Price
FROM airbnb
ORDER BY Price DESC;
SELECT 
    Name,
    Neighbourhood,
    Price
FROM airbnb_sorted
LIMIT 5;
```

###DDL Visualization

```python
%sql
SELECT 
    `Neighbourhood `,
    ROUND(AVG(Price), 2) AS avg_price
FROM airbnb
GROUP BY `Neighbourhood `
ORDER BY avg_price DESC;
```
