###DML 1 - Insert New Listings

```python
%sql
INSERT INTO airbnb (Name, `Property Type`, Price)
VALUES ('Sample Listing', 'Brooklyn', 150);
SELECT Name, `Property Type`, Price
FROM airbnb
WHERE Name = 'Sample Listing';
```

###DML 2 - Update Listings Price

```python
%sql
UPDATE airbnb
SET Price = Price + 20
WHERE `Neighbourhood ` = 'Queens';
SELECT Name, `Neighbourhood `, Price
FROM airbnb
WHERE `Neighbourhood ` = 'Queens'
LIMIT 5;
```

###DML 3 - Aggregation by Neighbourhood

```python
%sql
SELECT Name, Price
FROM airbnb
WHERE Price > 0;
SELECT `Property Type`, AVG(Price) AS avg_price
FROM airbnb
WHERE Price > 0
GROUP BY `Property Type`;
```

###DML 4 - High Price Filter

```python
%sql
SELECT 
    Name,
    Zipcode,
    Price
FROM airbnb
WHERE Price > 300
ORDER BY Price DESC
LIMIT 5;
```

###DML 5 - Count Total Listings

```python
%sql
SELECT 
    Price,
    COUNT(*) AS total_high_price
FROM airbnb
WHERE Price > 200
GROUP BY Price
ORDER BY total_high_price DESC;
```

###DML Visualization

```python
%sql
SELECT 
    `Zipcode`,`Price`,`Property Type`,`Neighbourhood `,
    ROUND(AVG(Price), 2) AS avg_price
FROM airbnb
WHERE Price > 100
GROUP BY `Zipcode`,`Price`,`Neighbourhood `,`Property Type`
ORDER BY avg_price DESC;
```
